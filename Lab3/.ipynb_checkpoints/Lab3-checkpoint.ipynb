# Intrinsic C & R Experiment / External Cap

```python
ExtCap = pd.read_excel("Lab3Data.xlsx",sheet_name="External Capacitor")
ExtCap['amplitude'] = ExtCap['amplitude']
C_k = 1.8e-9 #1.8nF
# Capacitance calculation
C_1 = C_k*(ExtCap['mseconds'][1]*1e-3/(ExtCap['mseconds'][0]*1e-3-ExtCap['mseconds'][1]*1e-3))
C_2 = ExtCap['amplitude'][0]*C_k/(ExtCap['amplitude'][1]-ExtCap['amplitude'][0]) 
C = np.mean([C_1,C_2])
Q_nc = ExtCap['amplitude'][1] * C
Q_c = ExtCap['amplitude'][0] * (C+C_k)
Q = np.mean([Q_nc,Q_c])
R_nc = ExtCap['mseconds'][1]*1e-3/C
R_c = ExtCap['mseconds'][0]*1e-3/(C+C_k)
R = np.mean([R_nc,R_c])
# R_k = R_nc*ExtCap['mseconds'][1]*1e-3/(ExtCap['mseconds'][1]*1e-3-R_nc*C)
#Q = C_k/(1/ExtCap['amplitude'][0]-1/ExtCap['amplitude'][1]) # Option 1 for Q calculation
# print("Q_1 = {:.4f}nC".format(Q*1e9))
#Q = ExtCap['amplitude'][1]*C # double check
print("C_1 = {:.4f}nF".format(C_1*1e9))
print("C_2 = {:.4f}nF".format(C_2*1e9))
print("C = {:.4f}nF".format(C*1e9))
print("Q_nc = {:.4f}nC".format(Q_nc*1e9))
print("Q_c = {:.4f}nC".format(Q_c*1e9))
print("Q = {:.4f}nC".format(Q*1e9))
print("R_nc = {:.4f}megaohms".format(R_nc*1e-6))
print("R_c = {:.4f}megaohms".format(R_c*1e-6))
print("R = {:.4f}megaohms".format(R*1e-6))
ExtCap
```

# Counting Stat Experiment

```python
countData = pd.read_excel("Lab3Data.xlsx",sheet_name="Counting Statistics")
lmba = np.mean(countData['Rate'])
def P100(m):
    return 100*(np.exp(-lmba)*lmba**m) / scipy.special.factorial(m)
print(np.mean(countData['Rate']))
# print(countData['Rate'].value_counts().index)
xnew = np.linspace(0,22,num=100)
yerror = np.sqrt(lmba)
chi_sqrd = np.sum((countData['Rate'].value_counts()-P100(countData['Rate'].value_counts().index))**2/yerror**2)
print("chi^2 = ", chi_sqrd, " <=", np.count_nonzero(countData['Rate'].value_counts().values))
rd_chi = chi_sqrd/7 #(np.count_nonzero(countData['Rate'].value_counts().values)-2)
#p = Beta * delta tf
fig, ax = plt.subplots()
countData['Rate'].value_counts().sort_index(ascending=True).plot(ax=ax, kind='bar', width=1, alpha=0.7, edgecolor='black', ls='-',label='')
ax.errorbar(countData['Rate'].value_counts().index-5, countData['Rate'].value_counts().values, yerr=yerror, capsize=4, ecolor='red', elinewidth=1, markeredgewidth=1, linestyle='None')
ax.plot(xnew-5, P100(xnew), color='green', ls='--', label='Predicted Poisson\nDistribution $P(m) = e^{-\lambda} \dfrac{\lambda^m}{m!}$')
ax.axvline(lmba-5, color='blue', ls='--')
ax.annotate(text='$\lambda = {:.2f}$'.format(lmba), xy=(lmba-5.2, 2), xytext=(4,4), rotation=-90, textcoords="offset points", fontsize=12, color='blue')
# ax.plot(xnew-5, P100(xnew)+np.sqrt(lmba), color='green', ls='--', label='Upper Bound')
# ax.plot(xnew-5, P100(xnew)-np.sqrt(lmba), color='red', ls='--', label='Upper Bound')
ax.plot([],[],' ',label='$\~\chi^2 = {:.2f} \quad(\leq 1) \quad\checkmark$'.format(rd_chi))
ax.plot([],[],' ',label='Confidence Value\n'+r'$\alpha \approx 0.43 \quad(\geq 0.1) \quad\checkmark$')
ax.set_title('Frequency Distribution $F(m)$ vs. $m$ Count Rate', fontsize=14)
ax.set_xlabel('Count Rate $m$', fontsize=12)
ax.set_ylabel('Frequency of Count Rate $F(m)$', fontsize=12)
ax.legend(loc='best',fontsize=14)
plt.show()
countData
```

# Dead Time 1 & 2 Experiments

```python
# has gamma ray and beta ray sides
deadData1 = pd.read_excel("Lab3Data.xlsx",sheet_name="Dead Time 1")
deadData2 = pd.read_excel("Lab3Data.xlsx",sheet_name="Dead Time 2")
deadData2['counts_per_sec'] = deadData2['counts']/deadData2['seconds']
# print(deadData2)
n_a = deadData1['counts_per_sec'][0]
n_b = deadData1['counts_per_sec'][1]
n_ab = deadData1['counts_per_sec'][2]
tau_p = 1/n_ab - (1/n_ab)*np.sqrt(1-(n_ab*(n_a+n_b-n_ab)/(n_a*n_b)))
tau = (n_a+n_b-n_ab)/(2*n_a*n_b)
print("tau = {:.4f}ms".format(tau*1e3))
print("tau_p = {:.4f}ms".format(tau_p*1e3))
deadData1
```

# Lead Absorption Experiment

```python
leadData = pd.read_excel("Lab3Data.xlsx",sheet_name="Lead Shielding Experiment")
leadData = leadData[leadData.index != 1] # exclude outlier 1
leadData = leadData[leadData.index != 0] # exclude outlier 0

leadData['thickness'] = leadData['mass_kg'] / (leadData['height_m']*leadData['width_m']) #rho*x = M / a (kg/m^2)
leadData['thickness'] = leadData['thickness']/11.343
leadData['count_per_sec'] = leadData['count']/leadData['time_sec']
leadData['logrhox_bg'] = np.log(leadData['count_per_sec'])
scale = 1
xnew = np.linspace(0,np.max(leadData['thickness']),num=200) #np.min(leadData['thickness'])
# print(np.linspace(np.min(leadData['thickness']),100,num=100))
# ERROR PROPAGATION
leadData['error_counting'] = np.sqrt(leadData['count_per_sec'])
leadData['error_thickness'] = np.sqrt(\
    (0.001/(leadData['height_m']*leadData['width_m']))**2+\
    (0.00001*leadData['mass_kg']/(leadData['height_m']**2*leadData['width_m']))**2+\
    (0.00001*leadData['mass_kg']/(leadData['height_m']*leadData['width_m']**2))**2\
)
leadData['error_ln'] = np.abs(leadData['error_counting']-count_rate_error)/np.abs(leadData['count_per_sec']-count_rate)
leadData[np.isnan(leadData)] = 0

# PLOT 4
fig, ax3 = plt.subplots()

# scipy.optimize.leastsq()
# [[I,u,B], c] = scipy.optimize.leastsq(func, xdata, intensity, [0.,0.,0.], err)
# perr = np.sqrt(np.diag(c))

m2, b2 = np.polyfit(leadData['thickness'],leadData['logrhox_bg'],1)
chi_sqrd = np.sum((leadData['logrhox_bg']-(m2*leadData['thickness']+b2))**2/leadData['error_ln']**2)
rd_chi = chi_sqrd/(len(leadData['thickness'])-2)
ax3.scatter(leadData['thickness'],leadData['logrhox_bg'])
print("m =",-m2,"b =", b2)
print("mu = ", -m2*10)
print("Calculated I_o =", np.exp(b2))
print("Measured I_o =", leadData['count_per_sec'][min(leadData['count_per_sec'].index)])
print("Chi^2 =", chi_sqrd, "~Chi^2 = ", rd_chi, "Degrees of Freedom =",len(leadData['thickness'])-2)

# PLOT 4 STUFF
ax3.plot(leadData['thickness'], m2*leadData['thickness']+b2, color='red', ls='-.', label='y = {:.5f}x + {:.5f}'.format(m2,b2))
ax3.plot([],[],' ',label='$\~\chi^2 = {:.2f} \quad(\leq 1)$'.format(rd_chi))
ax3.errorbar(leadData['thickness'], leadData['logrhox_bg'], yerr=leadData['error_ln'], xerr=leadData['error_thickness'], capsize=4, ecolor='red', elinewidth=1, markeredgewidth=1, linestyle='None')
ax3.set_title(r'Natural Logarithm of Thickness (no Background) $ln(I(x_i)-B)$ vs. Thickness $\rho x $', fontsize=14)
ax3.set_ylabel(r'Ln of Thickness w/o bg   $ln(I_o-B)$ ', fontsize=12)
ax3.set_xlabel(r'Thickness $x$ / $mm$', fontsize=12)
ax3.legend(loc='best',fontsize=14)
ax3.plot()
plt.show()
# MODEL
# ln(I_i - B) = ln(I_o)-mu*x
#      y      =    A   + B*x
# B != mu (dont trust!)
I_o =  np.exp(b2)
mu = -m2
def I(x):
    return I_o*np.exp(-mu*x)
# PLOT 5
chi_sqrd2 = np.sum((leadData['count_per_sec']-(I(leadData['thickness'])))**2/leadData['error_counting']**2)
rd_chi2 = chi_sqrd2/(len(leadData['thickness'])-2)
print('PLOT2')
print("Chi^2 =", chi_sqrd2, "~Chi^2 = ", rd_chi2, "Degrees of Freedom =",len(leadData['thickness'])-2)
fig, ax2 = plt.subplots()
ax2.scatter(leadData['thickness'],leadData['count_per_sec']) # x is in mm
ax2.plot(xnew,I(xnew),color='orange', label=r'Calculated Curve from $\mu$ = '+'{:.3f} '.format(mu*10)+'$cm^{-1}$')
ax2.plot([],[],' ',label='$\~\chi^2 = {:.2f} \quad(\leq 1)$'.format(rd_chi2))
ax2.errorbar(leadData['thickness'], leadData['count_per_sec'], xerr=leadData['error_thickness'], yerr=leadData['error_counting'], capsize=4, ecolor='red', elinewidth=1, markeredgewidth=1, linestyle='None')
ax2.set_title(r'Count Rate $m$ vs. Thickness $x$', fontsize=14)
ax2.set_ylabel('Count Rate $m$', fontsize=12)
ax2.set_xlabel(r'Thickness $x$ / $mm$', fontsize=12)
ax2.legend(loc='best',fontsize=14)
plt.show()

Table5 = pd.DataFrame()
Table5['energy_MeV'] = np.array([0.1022,0.1277,0.1703,0.2554,0.3405,0.4086,0.5108,0.6811,1.022,1.362,1.533,2.043,2.633,3.065,4.086,5.108,6.130,10.22,15.32,25.54])
Table5['mu_inverse_cm'] = np.array([59.9,33.6,16.4,6.31,3.39,2.42,1.68,1.16,0.771,0.620,np.NaN,0.499,np.NaN,np.NaN,0.469,0.491,np.NaN,0.607,0.698,0.835])
fig, ax4 = plt.subplots()
i = 5
j = 15
ax4.scatter(Table5['mu_inverse_cm'][i:j],Table5['energy_MeV'][i:j], label='Table 5 Data Points') # x is in 1/cm
# ax4.plot(leadData['thickness'],I(leadData['thickness']),color='orange')
# ax4.errorbar(leadData['thickness'], leadData['count_per_sec'], yerr=leadData['error_counting'], xerr=leadData['error_thickness'], capsize=4, ecolor='red', elinewidth=1, markeredgewidth=1, linestyle='None')
ax4.axhline(0.662, ls='-.', color='red')
ax4.axvline(Table5['mu_inverse_cm'][7], ls='-.', color='green')
ax4.annotate(text=r'$\gamma$  (0.662 MeV)', xy=(1.25, 0.662), xytext=(4,4), textcoords="offset points", fontsize=12, color='red')
ax4.annotate(text=r'$\mu$  (1.16 $cm^{-1}$)', xy=(1.16, 1), xytext=(4,4), textcoords="offset points", fontsize=12, color='red')
ax4.set_title(r'Absorbtion Coefficient $\mu$ vs. Photon Energy $E$', fontsize=14)
ax4.set_ylabel('Photon Energy $E$ / $MeV$', fontsize=12)
ax4.set_xlabel(r'Absorbtion Coefficient $\mu$ / $cm^{-1}$', fontsize=12)
ax4.legend(loc='best',fontsize=14)
plt.show()
Table5
# leadData
```
